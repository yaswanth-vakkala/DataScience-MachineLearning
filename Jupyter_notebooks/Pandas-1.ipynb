```python
# Cleaning data - casting datatypes and handling missing values
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
pdf2 = pd.DataFrame(people)
```

```python
pdf2
```

```python
pdf2.dropna()
```

```python
pdf2.dropna(axis='index', how='any') #default args #drops all rows with missing values
# axis can be index or columns , drop rows if missing values or drop columns if missing values
# how is criteria used to drop row or column. how = any means drops rows/columns with any missing values
```

```python
pdf2.dropna(axis='index', how='all') # drop rows if all values are missing
```

```python
pdf2.dropna(axis='columns', how='all')
```

```python
pdf2.dropna(axis='columns', how='any')
```

```python
pdf2.dropna(axis='index', how='any', subset=['email']) # drop missing values in email rows
# same result even if we put how as all
```

```python
pdf2.dropna(axis='index', how='all', subset=['last', 'email'])
```

```python
pdf2.replace('NA', np.nan, inplace=True) # replacing custom missing values
pdf2.replace('Missing', np.nan, inplace=True) # replacing custom missing values
```
