```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D , MaxPool2D , Flatten , Dropout
from keras.preprocessing.image import ImageDataGenerator
import os
```

```python
print(os.listdir('../data/OCT/'))
print(os.listdir('../data/OCT/train'))
print(os.listdir('../data/OCT/test'))
```

```python
root_dir = '../data/OCT'
train_dir = os.path.join(root_dir, 'train')
test_dir = os.path.join(root_dir, 'test')
train_cnv_dir = os.path.join(train_dir, 'CNV')
train_dme_dir = os.path.join(train_dir, 'DME')
train_drusen_dir = os.path.join(train_dir, 'DRUSEN')
train_normal_dir = os.path.join(train_dir, 'NORMAL')
test_cnv_dir = os.path.join(test_dir, 'CNV')
test_dme_dir = os.path.join(test_dir, 'DME')
test_drusen_dir = os.path.join(test_dir, 'DRUSEN')
test_normal_dir = os.path.join(test_dir, 'NORMAL')
```

```python
igen_train = ImageDataGenerator(
    rescale = 1/255,
    validation_split=0.2
)
igen_validation = ImageDataGenerator(rescale = 1/255)
igen_test = ImageDataGenerator(rescale = 1/255)
```

```python
train = igen_train.flow_from_directory(train_dir, 
                                            target_size=(150, 150),
                                            subset = 'training',
                                            batch_size = 64,
                                            color_mode='grayscale',
                                            class_mode = 'categorical')

validation = igen_train.flow_from_directory(train_dir, 
                                            target_size=(150, 150),
                                            subset = 'validation',
                                            batch_size = 64,
                                            color_mode='grayscale',
                                            class_mode = 'categorical')

test = igen_test.flow_from_directory(test_dir, 
                                            target_size=(150, 150),
                                            batch_size = 64,
                                            color_mode='grayscale',
                                            class_mode = 'categorical')
```

```python
train.class_indices
```

```python
train.image_shape
```

```python
modelb = Sequential()
modelb.add(Flatten())
modelb.add(Dense(256, input_shape=(22500,), activation = 'relu'))
modelb.add(Dense(128, activation = 'relu'))
modelb.add(Dense(64, activation = 'relu'))
modelb.add(Dense(4, activation = 'sigmoid'))
```

```python
modelb.compile(optimizer = 'adam', loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
modelb_history = modelb.fit(train, epochs=20, validation_data = validation)
```

```python
modelb_history.history.keys()
```

```python
loss, acc = modelb.evaluate(test)
print("Loss =",loss)
print("Accuracy =",acc*100, "%", sep="")
```

```python
modelb.save('modelb.h5')
modelb.save('modelb.keras')
```

```python
plt.plot(modelb_history.history['loss'])
plt.plot(modelb_history.history['val_loss'])
plt.legend(['Training', 'Validation'])
plt.title('Training and Validation losses')
plt.xlabel('No of epochs')
```

```python
plt.plot(modelb_history.history['accuracy'])
plt.plot(modelb_history.history['val_accuracy'])
plt.legend(['Training', 'Validation'])
plt.title('Training and Validation Accuracy')
plt.xlabel('No of epochs')
```

```python
import pickle

with open('modelbHistoryDict.pkl', 'wb') as mbfile:
    pickle.dump(modelb_history.history, mbfile)
```

```python
model1 = Sequential()

model1.add(Conv2D(32, (5,5), padding='SAME', activation='relu', input_shape=(150,150,1)))
model1.add(MaxPool2D(pool_size=(2, 2)))
model1.add(Dropout(0.5))

model1.add(Conv2D(64, (5,5), padding='SAME', activation='relu'))
model1.add(MaxPool2D(pool_size=(2, 2)))
model1.add(Dropout(0.5))

model1.add(Flatten())

model1.add(Dense(256, activation='relu'))
model1.add(Dropout(0.5))
model1.add(Dense(4, activation='sigmoid'))
model1.summary()
```

```python
model1.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model1_history = model1.fit(train, epochs = 12, validation_data = validation)
```

```python
loss, acc = model1.evaluate(test)
print("Loss =",loss)
print("Accuracy =",acc*100, "%", sep="")
```

```python
model1.save('model1.h5')
model1.save('model1.keras')
```

```python
model2 = Sequential()

model2.add(Conv2D(32, (5,5), padding='SAME', activation='relu', input_shape=(150,150,1)))
model2.add(MaxPool2D(pool_size=(2, 2)))
model2.add(Dropout(0.4))

model2.add(Conv2D(64, (5,5), padding='SAME', activation='relu'))
model2.add(MaxPool2D(pool_size=(2, 2)))
model2.add(Dropout(0.4))

model2.add(Conv2D(128, (5,5), padding='SAME', activation='relu'))
model2.add(MaxPool2D(pool_size=(2, 2)))
model2.add(Dropout(0.4))

model2.add(Flatten())

model2.add(Dense(512, activation='relu'))
model2.add(Dropout(0.4))
model2.add(Dense(256, activation='relu'))
model2.add(Dropout(0.4))
model2.add(Dense(128, activation='relu'))
model2.add(Dropout(0.4))
model2.add(Dense(4, activation='sigmoid'))
model2.summary()
```

```python
model2.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['Accuracy', 'Precision', 'Recall'])
```

```python
model2_history = model2.fit(train, epochs = 20, validation_data = validation)
```

```python
loss, acc, precision, recall = model2.evaluate(test)
print("Loss = ",loss)
print("Accuracy = ",acc*100, "%", sep="")
```

```python
model2.save('model2.h5')
model2.save('model2.keras')
```
