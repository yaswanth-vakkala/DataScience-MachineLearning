```python
import numpy as np
from skimage import io, segmentation, color
from sklearn.cluster import KMeans
```

```python
image = io.imread('/content/drive/MyDrive/data_files/images/imgseg.jpg')
```

```python
lab_image = color.rgb2lab(image)
```

```python
h,w,d = lab_image.shape
pixels = np.reshape(lab_image, (h*w,d))
```

```python
kmeans = KMeans(n_clusters=5)
labels = kmeans.fit_predict(pixels)
```

```python
seg_image = np.reshape(labels, (h,w))
```

```python
io.imshow(image)
io.show()
io.imshow(seg_image)
io.show()
```
